```python
def process_meme(input_path, output_path = "processed_week1.jpg"):
  im = Image.open(input_path)
  print("Original image: ")
  display(im)
  im_transf = im.convert('L')
  w, h = im.size
  im_transf = im_transf.resize((w//2, h//2))
  im_transf.save("processed_week1.jpg")
  print("Original width:", w)
  print("Original height:", h)
  display(im_transf)
  return w, h, im_transf
```

```python
w, h, img = process_meme("dog.jpg")
```
